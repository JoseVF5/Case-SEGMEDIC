Será preciso mudar o data type de paciente_id, profissional_id, valor_pago e data_atendimento porque não tem sentido considerar as paciente e profissional como float sendo idenficadores e valor pago realizar a substituição da vírgula por ponto para mudar o data type para fazer sentido com a coluna. Além da coluna data_atendimento não estar com o type correto o que dificultaria para montar um filtro com essa coluna

```python
df_limpo["paciente_id"] = df_limpo["paciente_id"].astype(int)

df_limpo["idprofissional"] = df_limpo["idprofissional"].astype(int)

df_limpo["valor_pago"] = df_limpo["valor_pago"].str.replace(',', '.').astype(float)
```

```python
df_limpo["data_atendimento"] = pd.to_datetime(df_limpo["data_atendimento"], dayfirst=True)
```

```python
df_limpo.rename(columns={'data_atendimento': 'Data_Atendimento',
                         'paciente_id': 'ID_paciente',
                         'nomeprocedimento': 'Nome_procedimento',
                         'grupoprocedimento': 'Categoria_procedimento',
                         'nomeunidade': 'Unidade',
                         'formapagamento':'Metodo_pagamento',
                         'idprofissional': 'ID_profissional'}, inplace = True)
```

```python
df_limpo.to_csv('Dados_limpos.csv', index=False, encoding='utf-8')
```
