2️⃣ Valor Atendido por Mês

- Qual o valor total atendido por mês?

```python
#houve um erro na mudança de data type da coluna. corrigir repetindo o processo de mudança de data type.

df_limpo['Data_Atendimento'] = pd.to_datetime(df_limpo['Data_Atendimento'])
```

```python
df_limpo.dtypes
```

```python
valor_mes_agrupado = df_limpo.groupby(pd.Grouper(key='Data_Atendimento', freq='ME'))['valor_pago'].agg(['sum', 'mean'])
```

ME = Mounth End

```python
valor_mes_agrupado = valor_mes_agrupado.rename(columns={
    'sum': 'Soma_Total', 
    'mean': 'Media_Mensal'
})
valor_mes_agrupado.head(25)
```

3️⃣ Quantidade de Atendimentos por Mês

- Quantos atendimentos ocorreram em cada mês?

```python
atendimento_mes = df_limpo.groupby(pd.Grouper(key='Data_Atendimento', freq='ME'))['ID_paciente'].count()
atendimento_mes.head(25)
```

4️⃣ Pacientes com Maior Frequência

Quais são os Top 5 pacientes com maior uso/frequência de atendimento?

```python
freq_pacientes = df_limpo.groupby(pd.Grouper(key='Data_Atendimento', freq='ME'))['ID_paciente'].value_counts().reset_index(name='Frequencia')
freq_pacientes_max = freq_pacientes.groupby('Data_Atendimento').head(1)
print(freq_pacientes_max.head(6))
```

utilizei o reset.index() para redefinir o índice do dataframe e adicionei o parâmetro 'name' para adicionar a coluna 'Frequencia'
